```python
import numpy as np
import torch_net
import torch
```

```python
N = 100
p = 10

X = np.random.randn(N,p)
Y = np.mean(np.abs(X), axis=1)
```

```python
model = torch_net.FCNet(
    input_size=p,
    output_size=1,
    hidden_sizes=[50,50,50],
    drop_rate=0.5
)
```

```python
def custom_loss(y1,y2):
    return torch.mean(torch.abs(y1-y2))

model.fit(X,Y, loss_fn=custom_loss, learning_rate=1e-3, epochs=50)
```

```python
yhat = model.predict(X)
yhat
```
